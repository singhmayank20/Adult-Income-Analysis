##### To Count No. of  different Categories for Categorical data.

```python
def value_counts():

    lists = ["workclass","education","marital_status","occupation","relationship","race","gender","native_country",
           "income"]
    
    lst=[]
    
    for item in lists:
        counts = data[item].value_counts().shape[0]
        lst.append({"Categorical data":item,"No. of  different Categories":counts})
        
    return pd.DataFrame(lst,columns=["Categorical data","No. of  different Categories"]) 
    
```

```python
value_counts()
```
